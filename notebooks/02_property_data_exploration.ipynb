```python
#Cell 1: Load property data
import pandas as pd
import matplotlib.pyplot as plt

#Load property boundary
# This section is commented out until you create the file.
# boundary = pd.read_csv('../data/raw/property_boundary.csv')
# print("Property boundary loaded:")
# print(boundary.head())
# print(f"\nShape: {boundary.shape}")

#Load observations
observations = pd.read_csv('../data/raw/observations.csv')
print("\n\nObservations loaded:")
print(observations.head())
```

```python
#Cell 2: Basic property statistics
print(f'Total observations recorded: {len(observations)}')
print(f"\nObservation categories:")
print(observations['category'].value_counts())
```

```python
#Cell 3: Simple visualization
if 'latitude' in observations.columns and 'longitude' in observations.columns:
    plt.figure(figsize=(10, 8))
    plt.scatter(observations['longitude'], observations['latitude'], c='red', s=100, alpha=0.6)
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.title('Property Observations Map')
    plt.grid(True, alpha=0.3)
    plt.show()
else:
    print("Add latitude/longitude columns to your observations.csv for mapping!")
```
