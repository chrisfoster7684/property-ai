```python
# Cell 3: Test plotting
import matplotlib.pyplot as plt
import numpy as np

# Create sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='sin(x)', linewidth=2)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('My First Plot - Environment Working!')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print("Visualization working! ✓")
```

```python
# Cell 4: Test weather API
import requests
import os
from dotenv import load_dotenv

# Load API key
load_dotenv()
api_key = os.getenv('OPENWEATHER_API_KEY')
print(f"My API key is: {api_key}") # Diagnostic print statement

# Your property location (update with your coordinates)
# Panola County, TX coordinates
lat = 32.1649  
lon = -94.3541 

# Get current weather
url = f"https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={api_key}&units=imperial"
response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    print("Weather API working! ✓")
    print(f"Current temperature: {data['main']['temp']}°F")
    print(f"Conditions: {data['weather'][0]['description']}")
else:
    print(f"Error: {response.status_code}")
```
